```python
import numpy as np

aveCores= np.mean(data['Number of Cores'])
aveThreads= np.mean(data['Number of Threads'])

print(f'Average Number of Threads: {aveThreads:.1f}')
print(f'Average Number of Cores: {aveCores:.1f}')
```
